```python
#ex 1
from gurobipy import Model, GRB
m = Model("Ex_1")
var_names = {"x_1", "x_2"}
x = [m.addVar(lb=0.0, vtype=GRB.CONTINUOUS, name = n) for n in var_names]
m.addConstr(3*x[0] + 2*x[1] <= 130)
m.addConstr(1*x[0] + 2*x[1] <= 100)

m.addConstr(x[0] <= 100)
m.addConstr(x[1] <= 150)

m.setObjective(1.5 * x[0] + 2 * x[1], GRB.MAXIMIZE)

m.optimize()
print("x_1", x[0]. X)
print("x_2", x[1]. X)
```

```python
#ex 3
from gurobipy import Model, GRB, quicksum
def preprocessing():
    i = 2
    j = 2
    a = [
        [3,2],
        [1,2]
    ]
    b = [130, 100]
    c = [1.5, 2.0]
    ub = [100, 150]
    return i, j, a, b, c, ub

def opt_model(i, j, a, b, c, ub):
    m = Model("Ex_3")

    x = [m.addVar(lb = 0, ub = ub[j], vtype = GRB.CONTINUOUS) for j in range(j)]

    m.setObjective(quicksum(c[j] *  x[j] for j in range (j)), GRB.MAXIMIZE)

    for i in range(i):
        m.addConstr(quicksum(a[i][j] * x[j] for j in range(j)) <=b[i])
    m.optimize()

    return m, x

def post_processing(m, x):

    if m.Status == GRB.OPTIMAL:
        print("obj:",  m.ObjVal)
        print("decision Variables:")
        for var in x:
            print(var.X)

        print("Shadow Prices")
        for constr in m.getConstrs():
            print(constr.Pi)

        print("Slack Values:")
        for constr in m.getConstrs():
            print(constr.Slack)

def main():
    i, j, a, b, c, ub = preprocessing()
    Model, x = opt_model(i,j, a , b, c, ub)
    post_processing(Model, x)

# optional
if __name__ == "__main__":
    main()
```
