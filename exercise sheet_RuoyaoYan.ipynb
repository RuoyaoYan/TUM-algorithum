#### Exercise 1.1
##### Objective function

$$
\max Z =
310(x_{1F}+x_{1C}+x_{1R})
+ 380(x_{2F}+x_{2C}+x_{2R})
+ 350(x_{3F}+x_{3C}+x_{3R})
+ 285(x_{4F}+x_{4C}+x_{4R})
$$

##### constraints

###### weight
$$
\begin{aligned}
x_{1F}+x_{2F}+x_{3F}+x_{4F} &\le 10 \quad &&\\
x_{1C}+x_{2C}+x_{3C}+x_{4C} &\le 16 \quad &&\\
x_{1R}+x_{2R}+x_{3R}+x_{4R} &\le 8  \quad &&
\end{aligned}
$$

###### volume
$$
\begin{aligned}
480x_{1F} + 650x_{2F} + 580x_{3F} + 390x_{4F} &\le 6800 \quad &&\\
480x_{1C} + 650x_{2C} + 580x_{3C} + 390x_{4C} &\le 8700 \quad &&\\
480x_{1R} + 650x_{2R} + 580x_{3R} + 390x_{4R} &\le 5300 \quad &&
\end{aligned}
$$

###### availability 

\begin{aligned}
x_{1F}+x_{1C}+x_{1R} &\le 18 \quad &&\\
x_{2F}+x_{2C}+x_{2R} &\le 15 \quad &&\\
x_{3F}+x_{3C}+x_{3R} &\le 23 \quad &&\\
x_{4F}+x_{4C}+x_{4R} &\le 12 \quad &&
\end{aligned}

###### balance
$$
\frac{x_{1F}+x_{2F}+x_{3F}+x_{4F}}{10}
=
\frac{x_{1C}+x_{2C}+x_{3C}+x_{4C}}{16}
=
\frac{x_{1R}+x_{2R}+x_{3R}+x_{4R}}{8}
$$

###### linear equations：
$$
\begin{aligned}
16(x_{1F}+x_{2F}+x_{3F}+x_{4F})
- 10(x_{1C}+x_{2C}+x_{3C}+x_{4C}) &= 0 \\
8(x_{1F}+x_{2F}+x_{3F}+x_{4F})
- 10(x_{1R}+x_{2R}+x_{3R}+x_{4R}) &= 0
\end{aligned}
$$
###### non-negativitiy 
$$
x_{ij} \ge 0 \quad \forall i \in \{1,2,3,4\},\; j \in \{F,C,R\}
$$

```python
from gurobipy import Model, GRB, quicksum
import gurobipy as grb
model = grb.Model()
x1F = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x1F")
x1C = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x1C")
x1R = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x1R")
x2F = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x2F")
x2C = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x2C")
x2R = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x2R")
x3F = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x3F")
x3C = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x3C")
x3R = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x3R")
x4F = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x4F")
x4C = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x4C")
x4R = model.addVar(vtype = grb.GRB.CONTINUOUS,name = "x4R")
model.addConstr(1*x1F+1*x2F+1*x3F+1*x4F <= 10)
model.addConstr(1*x1C+1*x2C+1*x3C+1*x4C <= 16)
model.addConstr(1*x1R+1*x2R+1*x3R+1*x4R <= 8)
model.addConstr(480*x1F+650*x2F+580*x3F+390*x4F <= 6800)
model.addConstr(480*x1C+650*x2C+580*x3C+390*x4C <= 8700)
model.addConstr(480*x1R+650*x2R+580*x3R+390*x4R <= 5300)
model.addConstr(1*x1F+1*x1C+1*x1R <= 18)
model.addConstr(1*x2F+1*x2C+1*x2R <= 15)
model.addConstr(1*x3F+1*x3C+1*x3R <= 23)
model.addConstr(1*x4F+1*x4C+1*x4R <= 12)
model.addConstr(x1F >= 0)
model.addConstr(x1C >= 0)
model.addConstr(x1R >= 0)
model.addConstr(x2F >= 0)
model.addConstr(x2C >= 0)
model.addConstr(x2R >= 0)
model.addConstr(x3F >= 0)
model.addConstr(x3C >= 0)
model.addConstr(x3R >= 0)
model.addConstr(x4F >= 0)
model.addConstr(x4C >= 0)
model.addConstr(x4R >= 0)
# balance part #######
WF = x1F + x2F + x3F + x4F
WC = x1C + x2C + x3C + x4C
WR = x1R + x2R + x3R + x4R
model.addConstr(16*WF == 10*WC)
model.addConstr(8*WF == 10*WR)
# balance#################

model.setObjective(310*x1F+310*x1C+310*x1R+380*x2F+380*x2C+380*x2R+350*x3F+350*x3C+350*x3R+285*x4F+285*x4C+285*x4R, sense = grb.GRB.MAXIMIZE)
model.setParam('outPutFlag', 0)
model.optimize()
print("maximized profit", f"{model.objVal:.3f}")

for i in model.getVars():
    print("tons carried", i.varName,"=", i.x)
```

#### Exercise 1.2
###### objective function:
$$
\max Z
= 10y_1 + 8y_2
- (x_{11} + x_{12})
- 1.5(x_{21} + x_{22})
- 3(x_{31} + x_{32})
$$

###### final produced amount：
$$
y_1 \ge 600 \\
y_2 \ge 700
$$

###### weight：
$$
x_{11} + x_{21} + x_{31} = y_1 \\
x_{12} + x_{22} + x_{32} = y_2
$$

###### ratio P1 ：
$$
0.4\,y_1 \le x_{11} \le 0.6\,y_1 \quad (0.4 \le x_{11}/y_1 \le 0.6) \\
0.1\,y_1 \le x_{21} \le 0.2\,y_1 \quad (0.1 \le x_{21}/y_1 \le 0.2) \\
0.2\,y_1 \le x_{31} \le 0.5\,y_1 \quad (0.2 \le x_{31}/y_1 \le 0.5)
$$


###### ratio P2 ：
$$
0.5\,y_2 \le x_{12} \le 0.6\,y_2 \quad (0.5 \le x_{12}/y_2 \le 0.6) \\
0.1\,y_2 \le x_{22} \le 0.4\,y_2 \quad (0.1 \le x_{22}/y_2 \le 0.4) \\
0.2\,y_2 \le x_{32} \le 0.3\,y_2 \quad (0.2 \le x_{32}/y_2 \le 0.3)
$$

###### llimition：
$$
x_{11} + x_{12} \le 2000 \\
x_{21} + x_{22} \le 1000 \\
x_{31} + x_{32} \le 500
$$

###### non-negativity：
$$
x_{ij} \ge 0,\quad y_j \ge 0
$$

```python
import gurobipy as grb

m = grb.Model()

x11 = m.addVar(name="x11")
x12 = m.addVar(name="x12")
x21 = m.addVar(name="x21")
x22 = m.addVar(name="x22")
x31 = m.addVar(name="x31")
x32 = m.addVar(name="x32")

#final
y1 = m.addVar(name="y1")
y2 = m.addVar(name="y2")


m.addConstr(y1 >= 600, name="min_P1") #demands
m.addConstr(y2 >= 700, name="min_P2") #demands
m.addConstr(x11 + x21 + x31 == y1, name="balance_P1")
m.addConstr(x12 + x22 + x32 == y2, name="balance_P2")
m.addConstr(0.4 * y1 <= x11)
m.addConstr(x11 <= 0.6 * y1)
m.addConstr(0.1 * y1 <= x21)
m.addConstr(x21 <= 0.2 * y1)
m.addConstr(0.2 * y1 <= x31)
m.addConstr(x31 <= 0.5 * y1)
m.addConstr(0.5 * y2 <= x12)
m.addConstr(x12 <= 0.6 * y2)
m.addConstr(0.1 * y2 <= x22)
m.addConstr(x22 <= 0.4 * y2)
m.addConstr(0.2 * y2 <= x32)
m.addConstr(x32 <= 0.3 * y2)

m.addConstr(x11 + x12 <= 2000, name="RM1_supply")
m.addConstr(x21 + x22 <= 1000, name="RM2_supply")
m.addConstr(x31 + x32 <= 500,  name="RM3_supply")

# results
revenue = 10 * y1 + 8 * y2
cost = (x11 + x12) + 1.5 * (x21 + x22) + 3 * (x31 + x32)
m.setObjective(revenue - cost, grb.GRB.MAXIMIZE)

m.setParam("OutputFlag", 0)
m.optimize()

print("Z =", m.objVal)
for v in m.getVars():
    print(v.VarName, v.X)

```

#### Exercise 1.3
##### Objective function
$$
\max Z =
95x_1 + 58x_2 + 52x_3 + 96x_4 + 48x_5 + 84x_6
+ 94x_7 + 56x_8 + 91x_9 + 70x_{10} + 54x_{11} + 61x_{12}
$$

##### Constraints 
###### weight
$$
180x_1 + 91x_2 + 131x_3 + 211x_4 + 142x_5 + 137x_6
+ 204x_7 + 176x_8 + 98x_9 + 183x_{10} + 157x_{11} + 108x_{12}
\le 900
$$

###### binary thingy
$$
x_i \in \{0,1\}, \quad i = 1,2,\dots,12
$$

```python
from gurobipy import Model, GRB

model = Model("1.3")
x1  = model.addVar(vtype=GRB.BINARY, name="x1")
x2  = model.addVar(vtype=GRB.BINARY, name="x2")
x3  = model.addVar(vtype=GRB.BINARY, name="x3")
x4  = model.addVar(vtype=GRB.BINARY, name="x4")
x5  = model.addVar(vtype=GRB.BINARY, name="x5")
x6  = model.addVar(vtype=GRB.BINARY, name="x6")
x7  = model.addVar(vtype=GRB.BINARY, name="x7")
x8  = model.addVar(vtype=GRB.BINARY, name="x8")
x9  = model.addVar(vtype=GRB.BINARY, name="x9")
x10 = model.addVar(vtype=GRB.BINARY, name="x10")
x11 = model.addVar(vtype=GRB.BINARY, name="x11")
x12 = model.addVar(vtype=GRB.BINARY, name="x12")


model.addConstr(
    180*x1 +  91*x2 + 131*x3 + 211*x4 +
    142*x5 + 137*x6 + 204*x7 + 176*x8 +     #weight
     98*x9 + 183*x10 + 157*x11 + 108*x12
    <= 900,
    name="capacity"
)

model.setObjective(
    95*x1 + 58*x2 + 52*x3 + 96*x4 +
    48*x5 + 84*x6 + 94*x7 + 56*x8 +
    91*x9 + 70*x10 + 54*x11 + 61*x12,
    sense=GRB.MAXIMIZE
)

model.setParam("OutputFlag", 0)   
model.optimize()

print(f"Optimal total value Z = {model.objVal:.1f}")

for var in model.getVars():
    if var.x > 0.5:   
        print(f"{var.varName} = {var.x}")


```

#### Exercise 1.4
##### objective function
$$
\max Z = \sum_{i=1}^{10} p_i x_i
$$

##### constraints 
###### time
$$
\sum_{i=1}^{10} t_i x_i \le 200
$$
###### capital
$$
\sum_{i=1}^{10} m_i x_i \le 5000
$$
$$
x_i \le d_i,\quad i = 1,\dots,10
$$
$$
x_i \ge 0,\quad i = 1,\dots,10
$$


```python

from gurobipy import Model, GRB   
import pandas as pd                                  

df = pd.read_csv("Data_Exercise 4.csv", header=None)

# generate var 
n_products = int(df.iloc[0, 1])
time_cap = float(df.iloc[1, 1]) #time
material_cap = float(df.iloc[2, 1]) #cap

I = list(range(n_products))
profit   = []   # 存 p_j
time = []   # 存 t_j
mat  = []   # 存 r_j
demand   = []   # 存 d_j


for j in I:
    col_index = 1 + j   
    profit.append(float(df.iloc[5, col_index]))
    time.append(float(df.iloc[6, col_index]))
    mat.append(float(df.iloc[7, col_index]))
    demand.append(float(df.iloc[8, col_index]))

# print(profit)
# print(time)
# print(mat)
# print(demand)

model = Model("exercise 1.4")
x = {} 

for j in I:
    var = model.addVar(lb=0.0,vtype=GRB.CONTINUOUS,name=f"x_{j+1}")   
    x[j] = var  

model.update()
#print(model)

objective_function = 0
for j in I:
    objective_function = objective_function + profit[j] * x[j]

model.setObjective(objective_function, GRB.MAXIMIZE)

lim_time = 0
for j in I:
    lim_time = lim_time + time[j] * x[j]

model.addConstr(lim_time <= time_cap, name="time_capacity")

lim_material = 0
for j in I:
    lim_material = lim_material + mat[j] * x[j]

model.addConstr(lim_material <= material_cap, name="material_capacity")

for j in I:
    model.addConstr(x[j] <= demand[j], name=f"demand_{j+1}")

model.setParam("OutputFlag", 0)
model.optimize()

if model.Status == GRB.OPTIMAL:
    optimal_x = [x[j].X for j in I]
    optimal_profit = model.ObjVal
    row_opt_unit   = 10   # "Optimal Production Units:"
    row_opt_profit = 11   # "Otimal Profit:"

    for j in I:
        df.iloc[row_opt_unit, 1 + j] = optimal_x[j]

    df.iloc[row_opt_profit, 1] = optimal_profit

    df.to_csv("Data_Exercise 4.csv", header=False, index=False)

    print("Optimal solution found.")
    print("Total profit:", optimal_profit)
    print("Result file saved to:", "Data_Exercise 4")
    print("Production plan:")
    for j in I:
        print(f"  Product {j+1}: {optimal_x[j]} units")
else:
    print("No optimal solution. Status code:", model.Status)
```
